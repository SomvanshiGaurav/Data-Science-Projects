```python
# Importing API
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading Data 
fifa = pd.read_csv(r"C:\Users\Admin\Desktop\CodeingNext\CourseMaterial\fifa19\data.csv")
fifa.head()
```

```python
# Selecting required column
fifa = fifa[['Name', 'Age', 'Nationality', 'Overall', 'Potential', 'Club', 'Value','Wage','Position']]
fifa
```

```python
# Max number of Player in Age group. 
plt.figure(figsize=(20,10))
plt.title('Players by Age')
plt.xlabel('No. of players')
plt.ylabel('Players Age')
sns.countplot(x="Age", data=fifa)
plt.show()
```

```python
# Max number of Player's Nationality
plt.figure(figsize=(20,10))
plt.title('Players by Nationality')
plt.xlabel('No.of players')
plt.ylabel('Players Nationality')
lplt=sns.countplot(x="Nationality", data=fifa,order=fifa['Nationality'].value_counts().sort_values(ascending=False).index)
lplt.set(xlim=(0,20))
```

```python
# Player's Nationality by Count
fifa["Nationality"].value_counts().head(20)
```

```python
# Player's Overall Performance
plt.figure(figsize=(20,10))
plt.title('Players by Overall')
plt.xlabel('No. of players')
plt.ylabel('Players Overall')
sns.countplot(x="Overall", data=fifa)
plt.show()
```

```python
# Player's Position
plt.figure(figsize=(20,10))
plt.title('Player Position')
plt.xlabel('No. of players')
plt.ylabel('Player Position')
sns.countplot(x="Position", data=fifa)
plt.show()
```

```python
def num_conversion(num):
    a = num.replace('€', '')
    if 'M' in a:
        a = float(a.replace('M', ''))*1000000
    elif 'K' in a:
        a = float(a.replace('K', ''))*1000
    return float(a)
fifa['Value'] = fifa['Value'].apply(lambda x: num_conversion(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: num_conversion(x))
fifa
```

```python
# Top 15 Players
Top_15 = fifa.sort_values(["Value"], ascending=False).head(15)
Top_15[["Name" ,"Age" ,"Nationality" ,"Club" ,"Position" ,"Value"]]
```

```python
def best_11(pos):
    fifa_tmp = fifa.copy()
    blist = []
    
    for i in pos:
        blist.append([i,fifa_tmp[fifa_tmp['Position'] == i]['Overall'].max(),
            fifa_tmp.loc[[fifa_tmp[fifa_tmp['Position'] == i]['Overall'].idxmax()]]['Name'].to_string(index = False),
            fifa_tmp.loc[[fifa_tmp[fifa_tmp['Position'] == i]['Overall'].idxmax()]]['Age'].to_string(index = False),
            fifa_tmp.loc[[fifa_tmp[fifa_tmp['Position'] == i]['Overall'].idxmax()]]['Club'].to_string(index = False),
            fifa_tmp.loc[[fifa_tmp[fifa_tmp['Position'] == i]['Overall'].idxmax()]]['Value'].to_string(index = False),
            fifa_tmp.loc[[fifa_tmp[fifa_tmp['Position'] == i]['Overall'].idxmax()]]['Wage'].to_string(index = False)
            ])
        fifa_tmp.drop(fifa_tmp[fifa_tmp['Position'] == i]['Overall'].idxmax(), inplace = True)
    return pd.DataFrame(np.array(blist).reshape(11,7),columns = ['Position', 'Name', 'Overall', 'Age', 'Club', 'Value', 'Wage']).to_string(index = False)
```

```python
# Best squad for 4-3-3
Best_433 = ['GK', 'RB', 'CB', 'CB', 'LB', 'CDM', 'CM', 'CAM', 'RW', 'ST', 'LW']
print (best_11(Best_433))
```
